## Graphics

### Function $ f(x) = e^{x} - \sin{x}$

#### Func graphic

```python
x = np.arange(-5, 1, 0.01)
y = np.exp(x) - np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

#### Interpolation graphic

```python
x_values = np.array([-5, -4.5, -4, -2, -1.5 -1, 0])
y_values = np.exp(x_values) - np.sin(x_values)
print(y_values)
```

```python
estimator = LagrangePolynomial(7)
estimator.fit(x_values, y_values)
```

```python
estimator.coef_
```

```python
plt.plot(x, np.poly1d(estimator.coef_)(x))
```

#### Linear regression graphic

```python
estimator = LinearRegression(10)
estimator.fit(x_values, y_values)
```

```python
print(estimator.coef_)
```

```python
cf = estimator.coef_
plt.plot(x, np.poly1d(estimator.coef_)(x))
```

### Function $ f(x) = 2^{x} - (\frac{1}{2}) ^ x  - 2$

### Interpolation graphic

```python
x = np.arange(-5, 1, 0.01)
y = 2 * x ** 2 - (0.5) ** x - 2
```

```python
plt.plot(x, y)
```

```python
x_values = np.array([-5, -4.5, -4, -2, -1.5 -1, 0])
y_values = np.array(2 * np.array(list([x ** 2 for x in x_values])) - np.array(list([0.5 ** x for x in x_values])) - 2)
```

```python
estimator = LagrangePolynomial(10)
estimator.fit(x_values, y_values)
```

```python
estimator.coef_
```

```python
plt.plot(x, np.poly1d(estimator.coef_)(x))
```

#### Linear regression graphic

```python
estimator = LinearRegression(10)
estimator.fit(x_values, y_values)
```

```python
estimator.coef_
```

```python
plt.plot(x, np.poly1d(estimator.coef_)(x))
```
